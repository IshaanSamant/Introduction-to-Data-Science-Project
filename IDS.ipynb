```python
import pandas as pd
import math
data=pd.read_csv("athlete_events.csv")
```

```python
data
```

```python
a=data['Age'].mean()
h=data['Height'].mean()
w=data['Weight'].mean()
```

```python
for i in range(1000):
    if(pd.isna(data['Age'][i])==True):
        data['Age'][i]=a
    if(pd.isna(data['Height'][i])==True):
        data['Height'][i]=h
    if(pd.isna(data['Weight'][i])==True):
        data['Weight'][i]=w
```

```python
data
```

```python
Age_Max=max(data['Age'])
Age_Min=min(data['Age'])
Height_Max=max(data['Height'])
Height_Min=min(data['Height'])
Weight_Max=max(data['Weight'])
Weight_Min=min(data['Weight'])
```

```python
for i in range(1000):
    data['Age'][i]=(data['Age'][i]-a)/(Age_Max-Age_Min)
    data['Weight'][i]=(data['Weight'][i]-w)/(Weight_Max-Weight_Min)
    data['Height'][i]=(data['Height'][i]-h)/(Height_Max-Height_Min)
```

```python
data
```

```python
age={}
for i in range(1000):
    if data['Age'][i] not in age:
        age[data['Age'][i]]=0
    age[data['Age'][i]]=age[data['Age'][i]]+1
```

```python
import matplotlib.pyplot as plt
k=0
for i in age:
    if k<11:
        plt.bar(age[i],i)
    k=k+1
plt.show()
```

```python
country={}
for i in range(1000):
    if data['Team'][i] not in country:
        country[data['Team'][i]]=0
    country[data['Team'][i]]+=1
labels=[]
val=[]
for i in country:
    labels.append(i)
    val.append(country[i])
plt.pie(val,labels=labels)
plt.axis('equal')
plt.show()
```

```python
import operator
sorted_x = sorted(country.items(), key=operator.itemgetter(1))
labels=[]
val=[]
for i in range(-11,0):
    labels.append(sorted_x[i][0])
    val.append(sorted_x[i][1])
plt.pie(val,labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
